Mengumpulkan atau mengimpor Data: Dapatkan dataset yang berisi variabel-variabel yang relevan.

```python
import pandas as pd

# Load the dataset
file_path = '/content/energy_dataset_BI.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

Variabel yang digunakan
1. Type_of_Renewable_Energy: Kode numerik yang mewakili jenis sumber energi terbarukan (1: Surya, 2: Angin, 3: Hidroelektrik, 4: Geotermal, 5: Biomassa, 6: Pasang Surut, 7: Gelombang).

5. Energy_Storage_Capacity_MWh: Kapasitas penyimpanan energi dalam megawatt-hours (MWh).
6. Storage_Efficiency_Percentage: Efisiensi sistem penyimpanan energi dalam persentase.
7. Grid_Integration_Level: Kode numerik yang mewakili tingkat integrasi ke jaringan (1: Terintegrasi Sepenuhnya, 2: Terintegrasi Sebagian, 3: Integrasi Minimal, 4: Mikrogrid Terisolasi).
8. Initial_Investment_USD: Biaya investasi awal dalam USD.
9. Funding_Sources: Kode numerik yang mewakili sumber pendanaan (1: Pemerintah, 2: Swasta, 3: Kemitraan Publik-Swasta).
10. Financial_Incentives_USD: Insentif keuangan dalam USD.
11. GHG_Emission_Reduction_tCO2e: Pengurangan emisi gas rumah kaca dalam ton setara CO2 (tCO2e).
12. Air_Pollution_Reduction_Index: Indeks pengurangan polusi udara.
13. Jobs_Created: jumlah pekerjaan yang tercipta.

```python
import matplotlib.pyplot as plt

# Adjusting the plot for the appropriate number of subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 20))

columns = [
    'Installed_Capacity_MW', 'Energy_Production_MWh', 'Energy_Consumption_MWh',
    'Energy_Storage_Capacity_MWh', 'Storage_Efficiency_Percentage', 'Initial_Investment_USD',
    'Financial_Incentives_USD', 'GHG_Emission_Reduction_tCO2e', 'Air_Pollution_Reduction_Index',
    'Jobs_Created'
]

for ax, column in zip(axes.flatten(), columns):
    data[column].plot.hist(ax=ax, bins=20, title=column)
    ax.set_xlabel(column)

# Remove empty subplots
for i in range(len(columns), len(axes.flatten())):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Convert non-numeric columns to numeric where applicable for better correlation analysis
data['Storage_Efficiency_Percentage'] = pd.to_numeric(data['Storage_Efficiency_Percentage'], errors='coerce')
data['Initial_Investment_USD'] = pd.to_numeric(data['Initial_Investment_USD'], errors='coerce')
data['Jobs_Created'] = pd.to_numeric(data['Jobs_Created'], errors='coerce')
data['GHG_Emission_Reduction_tCO2e'] = pd.to_numeric(data['GHG_Emission_Reduction_tCO2e'], errors='coerce')
data['Energy_Storage_Capacity_MWh'] = pd.to_numeric(data['Energy_Storage_Capacity_MWh'], errors='coerce')

# Convert 'Type_of_Renewable_Energy' to a numerical representation
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# Filling missing values before label encoding to avoid errors
data['Type_of_Renewable_Energy'] = data['Type_of_Renewable_Energy'].fillna('Unknown') # Or any suitable placeholder
data['Type_of_Renewable_Energy'] = le.fit_transform(data['Type_of_Renewable_Energy'])

# Recompute the correlation matrix after conversion, excluding non-numeric columns
correlation_matrix_updated = data.select_dtypes(include=['number']).corr() # Select only numeric columns

# Display the updated correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_updated, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Updated Correlation Matrix of Renewable Energy Project Factors')
plt.show()
```

Dari visualisasi data tersebut didapatkan pemahaman Matriks korelasi menunjukkan bahwa sebagian besar variabel memiliki korelasi yang sangat lemah (|r| <= 0.10) antara satu dengan yang lainnya. Hal ini menandakan tidak ada faktor yang dapat diidentifikasi sebagai pengaruh signifikan terhadap efisiensi proyek energi terbarukan berdasarkan korelasi dalam dataset ini.








Hal tersebut juga yang terjadi antara hubungan investasi awal berkorelasi dengan penciptaan lapangan kerja dalam proyek energi terbarukan

```python
# Calculate the average GHG emission reduction for each type of renewable energy
average_ghg_reduction = data.groupby('Type_of_Renewable_Energy')['GHG_Emission_Reduction_tCO2e'].mean().sort_values(ascending=False)

# Visualize the average GHG emission reduction for each type of renewable energy
plt.figure(figsize=(12, 6))
sns.barplot(x=average_ghg_reduction.index, y=average_ghg_reduction.values, palette='viridis')
plt.title('Average GHG Emission Reduction by Type of Renewable Energy')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Average GHG Emission Reduction (tCO2e)')
plt.xticks(rotation=45)
plt.show()

# Display the average GHG emission reduction values
average_ghg_reduction
```

Berdasarkan visualisasi tersebut Biomass memiliki pengurangan emisi gas rumah kaca rata-rata tertinggi, diikuti oleh Tidal dan Hydroelectric.
